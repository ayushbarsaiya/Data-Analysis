```python
def box_plot(df):
    i=1
    plt.figure(figsize = (20,50))
    for col in df.columns:
        plt.subplot(10,3,i)
        sns.boxplot(x = df[col], data = df,width = 0.5, fliersize = 3, linewidth = 1)
        i+=1   
```

```python
box_plot(df)
```

```python
def scatter(df):
    i=1
    plt.figure(figsize = (20,50))
    for col in df.columns:
        plt.subplot(10,3,i)
        sns.scatterplot(x = df[col],y=df["Amount"], data = df)
        i+=1   
```

```python
scatter(df)
```

```python
def line(df):
    i=1
    plt.figure(figsize = (20,50))
    for col in df.columns:
        plt.subplot(10,3,i)
        sns.lineplot(x = df[col],y=df["Amount"], data = df)
        i+=1   
```
