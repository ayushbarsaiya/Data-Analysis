```python
# this data is skewed, than not be a good idea to use mean  for replacing the missing values.
df["Product Base Margin"].mean()
```

```python
# if data is skewed, than good idea to use median  for replacing the missing values.
df["Product Base Margin"].median()
```

```python
# if data is skewed, than that is good idea to use mode also  for replacing the missing values.
df["Product Base Margin"].mode()
```

```python
# You can use mean value to replace the missing values in case the data distribution is symmetric.
# Consider using median or mode with skewed data distribution.
# 
```

```python
df["Product Base Margin"].fillna(df["Product Base Margin"].median(),inplace=True)
```
